```python
import requests
import os
import numpy as np
import json
from bs4 import BeautifulSoup
```

```python
with open(os.path.expanduser('~/Desktop/data_trentino_elinks.htm')) as f1:
    list_htm = f1.read().replace('\n', '')

list_soup = BeautifulSoup(list_htm, 'html.parser')
article_list = list_soup.find_all('article', {'class': 'col col-sm-3 moodboard__item--sm'})
article_links = set([article.find('a', {'role': 'link'})['href'] for article in article_list])
```

```python
events = []
for article_link in article_links:
    event_text = requests.get('https://www.visittrentino.info' + article_link).text
    event_soup = BeautifulSoup(event_text, 'html.parser')
    meta = event_soup.find('ul', attrs={ 'class': 'list-unstyled fz24 gallery__highlights__cat' })

    if meta is not None:
        metalist = []
#       get period, locality and typology details
        text = (meta.getText().replace('\n\n', ';').replace('\n', ':').replace('  ', '')[1:-1])
    
#       split and make separate lists for all three of them
        metalist = text.split(';')
        period = metalist[0]
        locality = metalist[1]
        typology = metalist[2]
        
#       get the value of period 
        period_list = []
        period_list = period.split(':')
        period = (period_list[1])

#       get the value of locality 
        locality_list = []
        locality_list = locality.split(':')
#         print(locality_list)
        locality = (locality_list[1])        
        
#       get the value of typology
        typology_list = []
        typology_list = typology.split(':')
        typology = (typology_list[1])           

        
        header = event_soup.find('div', attrs={ 'class': 'content-block mb60' })
        name = (header.h1.text)
        desc = (header.h2.text)
        content_el = event_soup.find('div', attrs={ 'class': 'wysiwyg mt30' })
        content = [p.getText().replace('\n\n', '') for p in content_el.find_all('p')]
        content1 = (str(content))
        sidebar = event_soup.find('div', { 'class': 'col-sm-3 col-sm-push-1 sidebar' })
        print(sidebar)
#         bars = sidebar.find_all('div', { 'class': 'mb30' })
# #         print(bars)
#         if len(bars) > 1:
#             d = bars[1].getText().replace('\n\n\n\n', ':').replace('\n', '').replace('  ', '')
# #             print(d)
#         if len(bars) > 2:
#             t = bars[2].getText().replace('\n\n\n\n', ':').replace('\n', '').replace('  ', '')
#             print(t)
        org = sidebar.find('address').find('span', {'class': 'strong'})
        if org is not None:
            org = (org.getText())
        else:
            org = ("Not specified")
            
        event = {'Name':name, 'Period': period, 'Locality':locality, 'Typology':typology, 'Description':desc,
                'Content':content1, 'Organizer': org}
        events.append(event)
        len(events)
# with open('events.json','w') as f:
#     json.dump(events,f,indent=4)
```
