```python
import requests
import os
import numpy as np
import json
from bs4 import BeautifulSoup
```

```python
with open(os.path.expanduser('~/Desktop/data_trentino_events.htm')) as f:
    data = f.read().replace('\n', '')
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
events = soup.find_all("div", { "class" : "wrapper" })
```

```python
result = []
for event in events:
    
    img = event.find("img").get("src")
    name = event.find(attrs={ 'itemprop' : 'name' })['content']
    start_date = event.find(attrs={ 'itemprop' : 'startDate' })['content']
#     duration = event.find(attrs={ 'itemprop' : 'duration' })['content']
    url = event.find(attrs={ 'itemprop' : 'url' })['content']
    place_temp = event.find("div", { "class" : "location" })
    print(place_temp)
    place = place_temp.text
    address = event.find(attrs={ 'itemprop' : 'address' })['content']
    latitude = event.find(attrs={ 'itemprop' : 'latitude' })['content']
    longitude = event.find(attrs={ 'itemprop' : 'longitude' })['content']
    
    inizio_temp = event.find("p", { "class" : "start" })
    if inizio_temp is not None:
        text ,begin = inizio_temp.text.split(":", 1)
    else:
        begin = 'No time specified'
        
    fine_temp = event.find("p", { "class" : "end" })
    if fine_temp is not None:
        text, end = fine_temp.text.split(":", 1)
    else:
        end = 'No time specified'    
        
    single ={'Image': img, 'Name':name,'Start Date':start_date,'URL':url,'Place':place,'Address':address,
            'latitude':latitude, 'longitude':longitude, 'Begin':begin, 'End':end}
    result.append(single)
#     print(single)
# with open('fleamarket.json','w') as f:
#     json.dump(result,f,indent=4)
len(result)

        
        
    
```
