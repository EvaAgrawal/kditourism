```python
import requests
import os
import json
from bs4 import BeautifulSoup
```

```python
with open(os.path.expanduser('~/Desktop/data_trentino_cinema.htm')) as f:
    data = f.read().replace('\n', '')
```

```python
soup = BeautifulSoup(data, 'html.parser')
links = soup.find_all("a", attrs = { "itemprop": "url" })
links = [link['href'] for link in links]
links = set(links)
```

```python
movies = []

for uri in links:
#     print(uri)
    r = requests.get('https://www.crushsite.it' + uri)
    spage = BeautifulSoup(r.text, 'html.parser')
    greybox = spage.find('div', attrs={ 'class' : 'colonnas-1-301ds' })
    name = greybox.find(attrs={ 'itemprop' : 'name' })['content']
    start_date = greybox.find(attrs={ 'itemprop' : 'startDate' })['content']
    end_date = greybox.find(attrs={ 'itemprop' : 'endDate' })['content']
#     print(greybox)

    for e in greybox.find_all('span', attrs={ 'class' : 'testo-boxgrigio-grassetto' }):
#         address = e.getText.find(attrs={'itemprop':'streetAddress'}))
#         print (address)
#     print(address)
        tag = (e.getText())
        print(tag)
#         for t in tag:
#             print(e.find_next_sibling().getText())
            
    desc = spage.find('div', attrs={ 'class' : 'colonna-1-5007-testoevi' })
    desc = desc.find('span', attrs={ 'class' : 'testo-boxgrigio-grassetto' })
#     print (desc)
    if desc is not None:
        desc = desc.getText().split('\n\t')
    else:
        desc = ("NO DESCRIPTION!")
    movie ={'Name':name,'Start Date':start_date,'End date':end_date, 'Description': desc}
    movies.append(movie)
#     print(movie)

# with open('movies.json','w') as f:
#     json.dump(movies,f,indent=4)
# len(movies)  
```
